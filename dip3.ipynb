```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
import skimage.util as sk


#Taking the image 
img = cv2.imread('Lena.png')
# convert it into gray scale image
grayScale_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#the cv2.resize() to down sample the image with 4 sizes
l = [0.5,0.25,0.125,0.0625]
img = [cv2.resize(grayScale_img,None, fx =i, fy = i) for i in l]

# Plot the original image and its down-sampled versions
plt.figure(figsize=(20, 10))
plt.subplot(1, 5, 1)
plt.imshow(grayScale_img,cmap='gray')
plt.title("Original Image")
for i, down_img in enumerate(img):
    plt.subplot(1, 5, i + 2)
    plt.imshow(down_img,cmap='gray')
    plt.title("Down-sampled Image {}".format(i+1))
plt.show()

#grayscale Lena image with 8 different intensity
l = [1,2,4,8,16,32,64,128]
img = [np.ceil(grayScale_img/i) for i in l]

# Plot the original image and its down-sampled versions
plt.figure(figsize=(20, 10))
for i, down_img in enumerate(img):
    plt.subplot(2, 4, i+1)
    plt.imshow(down_img,cmap='gray')
    plt.title("Reduced intensity Image {}".format(i+1))
plt.show()
```
